### Computation on Arrays: Broadcasting 

We saw in the previous section how NumPy’s universal functions can be used to vectorize operations and thereby remove slow Python loops. Another means of vectorizing operations is to use NumPy’s broadcasting functionality. Broadcasting is simply a set of rules for applying binary ufuncs (addition, subtraction, multiplication, etc.) on arrays of different sizes.



```python
import numpy as np
```

```python
a = np.array([0,1,2])
b = np.array([5,5,5])
a + b
```

```python
a + 5
```

* Broadcasting allows these types of binary operations to be performed on arrays of different sizes — for example, we can just as easily add a scalar (think of it as a zero-dimensional array)

 

```python
# can extend this to arrays of higher dimension
M = np.ones((3,3))
M
```

```python
# One-dimensional array (a) is stretched, or broadcast, across the second dimesion 
# in order to match the shape of M
M + a 
```

```python
a_1 = np.arange(3)
b_1 = np.arange(3)[:, np.newaxis]
print(a_1)
print(b_1)
```

```python
# stretched both a_1 and b_1 to match a common shape (Each index in latter b_1 is added 
# to a_1 one at a time producing 3 rows, 1 for each value in b_1 plus the one static a_1 row)
a_1 + b_1 
```

### Rules of Broadcasting

* Rule 1: If the two arrays differ in their number of dimensions, the shape of the one with fewer dimensions is padded with ones on its leading (left) side.
* Rule 2: If the shape of the two arrays does not match in any dimension, the array with shape equal to 1 in that dimension is stretched to match the other shape.
* Rule 3: If in any dimension the sizes disagree and neither is equal to 1, an error is raised.


```python
# Broadcasting example 1
M = np.ones((2,3))
a = np.arange(3)
print(M, '\n', a)
```

```python
# We see by rule 1 that the array "a" has fewer dimension, so we pad it on the left w/ones
print(M.shape, a.shape) 
# M.shape = (2,3), a.shape = (1,3)
a = a.reshape(1,3)
print(a, a.shape)
```

* By rule 2, we now see that the first dimension disagrees, so we stretch this dimension to match
* M.shape = (2,3) & a.shape(2,3)
* The shapes match, and we see that the final shape will be (2,3):

```python
print(M + a, (M+a).shape) 
```

* a is stretched (essentially duplicated) to the have it's index added to each similar index in the 2D Array
* a[0] + M[0][0] = 1 then a[0] + M[1][0] = 1
* Each of these addition happens for each element of the inner array for all of its' rows

```python
# Braodcasting example 2
a_r = np.arange(3).reshape((3, 1))
b = np.arange(3)
print(a_r, '\n', b)
print(a_r.shape, b.shape)
```

#### Rule 1 says we must pad the shape of b w/ones
* a.shape -> (3,1)
* b.shape -> (1, 3)

#### Rule 2 tells us that we upgrade each of these ones to match the corresponding size of the other array
* a.shape -> (3,3)
* b.shape -> (3,3)

```python
print(b[0] + a_r[0][0], b[0] + a_r[1][0], b[0] + a_r[2][0])
print(a_r[0][0] + b[0], a_r[0][0] + b[1], a_r[0][0] + b[2])
a_r + b
```

* So looks like each value for let's say b is added to each column value in a_r with each addition 
* Ex b[0] + a_r[0][0]  == 0

```python
# Broadcasting example 3
M = np.ones((3,2))
a = np.arange(3)
print(M.shape, a.shape)
```

#### Broadcasting in Practice
* Centering an Array

```python
np.random.seed(3)
X = np.random.random((10,3)) # random 0-1, 10 rows and 3 columns
X
```

```python
Xmean, Xmean_cols, Xmean_rows = X.mean(), X.mean(0), X.mean(1)
print("Mean of the entire set : {}".format(Xmean))
print("Mean for each column : {}".format(Xmean_cols))
print("Mean for each row : {}".format(Xmean_rows))
```

```python
# Test against slicing and columns
print(X[:, 0].mean(), X[:, 1].mean(), X[:, 1].mean())
```

#### Back to Centering!

```python
# we can center the X array by subtracting the mean (this is a broadcasting operation)
X_centered = X - X.mean(0)
print(X_centered)
print(X[1,0], X[1][0], X[1][0] - X.mean(0)[0], X_centered[1][0]) # same thing for X second row first column
```

```python
X_centered.mean(0) # To within-machine precision, the mean is now zero
```

### Boolean Arrays

```python
rng = np.random.RandomState(0)
x = rng.randint(10, size=(3,4))
x
```

```python
x < 6
```

```python
# how many values less than 6
np.count_nonzero(x < 6)
```

```python
np.sum(x < 6)
```

```python
# how many values less than 6 in each row
np.sum(x < 6, axis=1)
```

```python
# how many values less than 6 in each column
np.sum(x < 6, axis=0)
```

```python
# any values greater than 8
np.any(x > 8)
```

```python
# any values less than zero
np.any(x < 0)
```

```python
# are all values less than 10
np.all(x < 10)
```

```python
# all values equal to 6
np.all(x == 6)
```

```python
# are all values in each row less than 8
np.all(x < 8, axis=1)
```
